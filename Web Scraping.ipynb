```python
import pymongo
import requests
import time
import threading

#Give your server name $ password after momgodb://
conn=pymongo.MongoClient("mongodb://guvidm2:Admin123@ac-phzdj1b-shard-00-00.m2i4fsw.mongodb.net:27017,ac-phzdj1b-shard-00-01.m2i4fsw.mongodb.net:27017,ac-phzdj1b-shard-00-02.m2i4fsw.mongodb.net:27017/?ssl=true&replicaSet=atlas-vmdwu3-shard-0&authSource=admin&retryWrites=true&w=majority")


def fetch_and_upload_t1_mongodb():
    
    database=conn["Threading_data_fetch"]
    collection=database["satellite_position"]
    
    for i in range(3600):
        try:
            response=requests.get("http://api.open-notify.org/iss-now.json")
            response_json=response.json()
            collection.insert_one(response_json)
            print(response_json,time.strftime("%H:%M:%S"))
        except:
            print("ERROR",response.status_code)
        
        time.sleep(1)

def fetch_and_upload_t2_mongodb():
    
    database=conn["Threading_data_fetch"]
    collection=database["Crypto_ticker"]
    
    for i in range(3600):
        try:
            response=requests.get("https://www.bitstamp.net/api/v2/ticker/btcusd")
            response_json=response.json()
            collection.insert_one(response_json)
            print(response_json,time.strftime("%H:%M:%S"))
        except:
            print("ERROR",response.status_code)
        
        time.sleep(1)

        
if __name__=='__main__':
    
    
    thread1=threading.Thread(target=fetch_and_upload_t1_mongodb())
    thread2=threading.Thread(target=fetch_and_upload_t2_mongodb())
    
    thread1.start()
    thread2.start()
    
    thread1.join()
    print("Satellite position Upload complete!!!")
    
    thread2.join()
    print(" Upload complete!!!")
```
